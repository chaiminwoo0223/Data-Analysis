```python
import pandas as pd
exam = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/exam.csv')
exam
```

```python
exam.sort_values('math') # math 오름차순 정렬
```

```python
exam.sort_values('math',ascending = False) # math 내림차순 정렬
```

```python
exam.sort_values(['nclass','math']) # nclass, math 오름차순 정렬
```

```python
exam.sort_values(['nclass','math'],ascending = [True,False]) # nclass 오름차순, math 내림차순 정렬
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science']) # total 변수 추가
```

```python
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],      # total 추가
    mean = (exam['math'] + exam['english'] + exam['science']) / 3) # mean 추가
```
