```python
import pandas as pd
exam = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/midwest.csv') # 데이터 불러오기
exam.head()
```

```python
exam = exam.rename(columns = {'poptotal' : 'total','popasian' : 'asian'}) # poptotal을 total로 수정, popasian을 asian으로 수정
exam.head()
```

```python
exam['ratio'] = (exam['asian'] / exam['total']) * 100 # 전체 인구 대비 아시아 인구 백분율
exam
```

```python
exam['ratio'].plot.hist() # 히스토그램
```

```python
mean = exam['ratio'].mean() # 평균
print(mean)
```

```python
import numpy as np # 넘파이
exam['size'] = np.where(exam['ratio'] > mean,'large','small') # 평균 이상이면 large, 그렇지 않으면 small 부여
exam
```

```python
count_size = exam['size'].value_counts().sort_index() # 빈도표, 알파벳순으로 정렬
count_size
```

```python
count_size.plot.bar(rot = 0) # 빈도 막대 그래프
```
