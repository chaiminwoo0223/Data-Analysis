```python
import pandas as pd
exam = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/exam.csv')
exam
```

```python
exam['math'].mean()
```

```python
exam.agg(mean_math = ('math','mean')) # 수학 평균
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean')) # 반별 수학 평균
```

```python
exam.groupby('nclass',as_index = False).agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass')\
    .agg(mean_math   = ('math','mean'),
         sum_math    = ('math','sum'),
         median_math = ('math','median'),
         n           = ('nclass','count'))
```

```python
exam.groupby('nclass').mean() # 모든 변수의 평균
```

```python
mpg = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/mpg.csv')
mpg
```

```python
# 제조 회사 및 구동 방식별로 분리한 후, cty 평균 구하기
mpg.groupby(['manufacturer','drv'])\
           .agg(mean_cty = ('cty','mean'))
```

```python
# audi의 drv별 빈도
mpg.query('manufacturer == "audi"')\
   .groupby(['drv'])\
   .agg(n = ('drv','count')) 
```

```python
# chevrolet의 drv별 빈도
mpg.query('manufacturer == "chevrolet"')\
   .groupby(['drv'])\
   .agg(n = ('drv','count')) 
```

```python
mpg.groupby(['drv'])\
   .agg(n = ('drv','count')) # drv별 빈도
```

```python
mpg['drv'].value_counts() # value_counts()로 간단하게 빈도 구하기
```

```python
# Q. 제조 회사별로 'suv' 자동차의 도시 및 고속도로 합산 연비 평균을 구해 내림차순으로 정렬하고, 1 ~ 5위까지 출력하기
mpg.query('category == "suv"')\
   .assign(total = (mpg['hwy'] + mpg['cty'] / 2))\
   .groupby('manufacturer')\
   .agg(mean_tot = ('total','mean'))\
   .sort_values('mean_tot',ascending = False)\
   .head()
```
