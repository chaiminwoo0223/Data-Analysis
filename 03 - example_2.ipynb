```python
import pandas as pd
exam = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/exam.csv')
exam
```

```python
exam.query('nclass == 1') # 1반인 경우, 
```

```python
exam.query('nclass == 2') # 2반인 경우
```

```python
exam.query('nclass != 1') # 1반이 아닌 경우
```

```python
exam.query('nclass != 3') # 3반이 아닌 경우
```

```python
exam.query('math > 50') # 수학점수 50점 초과
```

```python
exam.query('math < 50') # 수학점수 50점 미만
```

```python
exam.query('math == 50') # 수학점수 50점
```

```python
exam.query('english >= 50') # 영어점수 50점 이상 
```

```python
exam.query('english <= 80') # 영어점수 50점 이하 
```

```python
exam.query('nclass == 1 & math >= 50') # 1반 and 수학점수 50점 이상
```

```python
exam.query('nclass == 2 & english >= 80') # 2반 and 영어점수 80점 이상
```

```python
exam.query('math >= 90 | english >= 90') # 수학점수 90점 이상 or 영어점수 90점 이상
```

```python
exam.query('math < 90 | english < 50') # 수학점수 90점 미만 or 영어점수 50점 미만
```

```python
exam.query('nclass == 1 | nclass == 3 | nclass == 5') # 1반 or 3반 or 5반
```

```python
exam.query('nclass in [1,3,5]') # 1반 or 3반 or 5반
```
