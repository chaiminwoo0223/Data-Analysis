```python
import pandas as pd
exam = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/exam.csv')
exam
```

```python
exam.sort_values('math') # math 오름차순 정렬
```

```python
exam.sort_values('math',ascending = False) # math 내림차순 정렬
```

```python
exam.sort_values(['nclass','math']) # nclass, math 오름차순 정렬
```

```python
exam.sort_values(['nclass','math'],ascending = [True,False]) # nclass 오름차순, math 내림차순 정렬
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science']) # total 변수 추가
```

```python
exam.assign(
    total = exam['math'] + exam['english'] + exam['science'],      # total 추가
    mean = (exam['math'] + exam['english'] + exam['science']) / 3) # mean 추가
```

```python
import numpy as np
exam.assign(test = np.where(exam['science'] >= 60, 'pass', 'fall'))
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science']).sort_values('total') # total 오름차순 정렬
```

```python
long_name = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/exam.csv')
long_name.assign(new = long_name['math'] + long_name['english'] + long_name['science']) # long_name 직접 입력
```

```python
long_name.assign(new = lambda x: x['math'] + x['english'] + x['science']) # lambda 이용하여, long_name 대신 x 입력
```

```python
exam.assign(total = lambda x: x['math'] + x['english'] + x['science'], # 앞에서 만든 total 활용하여, mean 만들기(lambda 활용)
            mean = lambda x: x['total'] / 3)
```
