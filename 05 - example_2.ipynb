```python
import pandas as pd
df_raw = pd.read_csv('/content/drive/MyDrive/수업/데이터 분석 입문/강의자료/Data/exam.csv')
df = df_raw.head(10)
df
```

```python
df['nclass'] # nclass 추출
```

```python
df['nclass'] == 1 # 1반 확인
```

```python
df[df['nclass'] == 1] # 1반만 추출
```

```python
df[df['math'] >= 80] # 수학점수가 80점 이상이면 추출
```

```python
df[(df['nclass'] == 1) & (df['math'] >= 50)] # 1반이면서, 수학점수가 50점 이상
```

```python
df[(df['math'] < 50) | (df['english'] < 50)] # 수학점수가 50점 미만이거나, 영어점수가 50점 미만
```
